# Imbalanced class classification problem!

#### About Data:
Output variable (desired target):
Y - has the client subscribed a term deposit? (binary: 'yes','no')

Input variables:
#### bank client data:
1 - age (numeric)

2 - job : type of job (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')

3 - marital : marital status (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)

4 - education (categorical: 'basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course','university.degree','unknown')

5 - default: has credit in default? (categorical: 'no','yes','unknown')

6 - housing: has housing loan? (categorical: 'no','yes','unknown')

7 - loan: has personal loan? (categorical: 'no','yes','unknown')

#### related with the last contact of the current campaign:

8 - contact: contact communication type (categorical: 'cellular','telephone') 

9 - month: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')

10 - day_of_week: last contact day of the week (categorical: 'mon','tue','wed','thu','fri')

11 - duration: last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

#### other attributes:

12 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)

13 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)

14 - previous: number of contacts performed before this campaign and for this client (numeric)

15 - poutcome: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')

#### social and economic context attributes

16 - emp.var.rate: employment variation rate - quarterly indicator (numeric)

17 - cons.price.idx: consumer price index - monthly indicator (numeric) 

18 - cons.conf.idx: consumer confidence index - monthly indicator (numeric) 

19 - euribor3m: euribor 3 month rate - daily indicator (numeric)

20 - nr.employed: number of employees - quarterly indicator (numeric)

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Label Encode:
def lbl_encode(dframe, list_of_column_name): # yes = 1, No = 0
    '''
    args: dataframe , list of columns to be label encoded
    returns: a data frame
    '''
    for i in list_of_column_name:
        lenc = LabelEncoder()
        dframe[[i]] = lenc.fit_transform(dframe[[i]])
    return dframe
```

```python
to_label_encode = ['housing', 'loan', 'default', 'month', 'y']  # try these: campaign
df = lbl_encode(df, to_label_encode) # yes = 1, No = 0

df.sample(3)
```

```python
to_onehot_encode = ['job', 'marital', 'education', 'contact', 'poutcome', 'month', 'age'] 

df = pd.get_dummies(
    data= df, 
    prefix_sep= '_', 
    columns= to_onehot_encode, 
    drop_first= True
)
df.head(5)
```

```python
# Scaling the columns:
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Based on the summary statitics from 'df2.describe().T' below are the column names 
# that need to be scaled because they have high variance and possibly will create 
# problem in the learning process for any algorthm.

column_to_scale = ['balance', 'duration']

# Min_Max_Scale:
def min_max_scaling(dframe, list_of_column_name): # return a value between 0 and 1
    '''
    args: dataframe , list of columns to be label encoded
    returns: a data frame
    '''
    for i in list_of_column_name:
        mms = MinMaxScaler()
        dframe[[i]] = mms.fit_transform(dframe[[i]])
    return dframe

# standard scaling:
def standard_scaling(dframe, list_of_column_name): # return a value between 0 and 1
    '''
    args: dataframe , list of columns to be label encoded
    returns: a data frame
    '''
    for i in list_of_column_name:
        sc = StandardScaler()
        dframe[[i]] = sc.fit_transform(dframe[[i]])
    return dframe


df2 = min_max_scaling(df, column_to_scale)
print(df2.describe().T)

df3 = standard_scaling(df, column_to_scale)
print(df3.describe().T)

del(df)
```

### Lets split the data in train and test set and apply ML algorithm for classification

```python
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from imblearn.over_sampling import SMOTE

y = df2.y.values
df2 = df2.drop('y', axis=1)
X = df2.values

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.25, 
    random_state=0
)

print(X_train.shape, y_train.shape)

# sm = SMOTE(random_state=12, ratio = 1.0)
# X_train, y_train = sm.fit_sample(X_train, y_train)

# print(X_train.shape, y_train.shape)
```

### Lets Apply Grid Search to get Best parameters for our ML Algorithms

```python
# Importing Metrics and helper classes
import time
from sklearn.metrics import classification_report, roc_auc_score, recall_score, accuracy_score, f1_score
from sklearn.model_selection import cross_val_score, GridSearchCV

# Importing Machine Learning Classifiers from SKLEARN Library
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
# from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
# from sklearn.linear_model import SGDClassifier

### Algorithms to try! ###
# lof , KNN, RF, ADA (DT), LR, Isolation Forest, 1 class SVM, Multinomial Naive Bayes, Gaussian NB,
# Bernouli Naive Bayes, MLP, SVM


estimator = [
    LogisticRegression(max_iter=300),
    DecisionTreeClassifier(class_weight= 'balanced'),
    RandomForestClassifier(n_jobs = -1),
#     MLPClassifier(),
    GradientBoostingClassifier(),
    AdaBoostClassifier(random_state =0)
#     SVC(kernel="linear", C=0.025, n_jobs = -1,),
#     SVC(gamma=2, C=1, n_jobs = -1,),
#     GaussianProcessClassifier(1.0 * RBF(1.0), n_jobs = -1,)
]

names = [
        "Logistic_Regression",
        "Decision_Tree", 
        "Random_Forest",
#         "MLPClassifier",
        "GradientBoostingClassifier",
        "AdaBoostClassifier",
#         "SVC_linear",
#         "SVC",
#         "GaussianProcessClassifier",
#         "QuadraticDiscriminantAnalysis"
        ]

parameters = {
        "Logistic_Regression":{
                        'solver': ['liblinear'],
                        'class_weight': ['balanced']
                        },
        "Decision_Tree": {
                        "max_depth": [15, 32], 
                        "criterion": ["entropy"]
                        },
        "Random_Forest": {
                        'n_estimators': [50, 80], 
                        'max_features': ['log2'], 
                        'criterion': ["gini"]
                        },
#         "MLPClassifier": {
#                         'hidden_layer_sizes': [(16,32,64)],
#                         'activation' : ['tanh'],
#                         'solver': ['adam']
#                         },
        "GradientBoostingClassifier" : {
                        'loss' : ['exponential'],
                        'max_features': ['log2'], 
                        'max_depth' : [30]
                        },
        "AdaBoostClassifier" : {
                        'n_estimators':[300]
                        }
#         "SVC_linear": {
#                         },
#         "SVC": {
#                         },
#         "GaussianProcessClassifier": {
#                         },
#         "QuadraticDiscriminantAnalysis":{
#                         }
        }


results = pd.DataFrame(columns=[
                            'Algorithm', 'Best_Parameters', 'Best_Estimator', 'recall', 'f1_score',
                            'accuracy', 'roc_auc', 'Test_Score'])

for name, algo in zip(names, estimator):
        now = time.time()
        cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
        clf = GridSearchCV(
                    estimator = algo,
                    param_grid = parameters[name],
                    n_jobs = -1,
                    scoring = 'roc_auc',
                    cv = cv
                    )
            
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        
        results = results.append({
                'Algorithm': name, 
                'roc_auc': roc_auc_score(y_test,clf.predict(X_test)), 
                'accuracy': accuracy_score(y_test,clf.predict(X_test)), 
                'recall': recall_score(y_test,clf.predict(X_test)), 
                'f1_score' : f1_score(y_test,clf.predict(X_test)), 
                'Best_Parameters': clf.best_params_,
                'Best_Estimator': clf.best_estimator_,
                'Test_Score' : score
            }, ignore_index=True)
        
        later = time.time()
        print('Time taken for', name,'metric is: ',int(later - now))
```
